#### - letalidad vs dias de estada:
 Pacientes con más días de estadía podrían presentar mayor letalidad; veamos cuan cierta es la aseveración...

```python
let_vs_dias = df_indicadores[["LETALIDAD", "PROMEDIO_DIAS_ESTADA"]].groupby("LETALIDAD")
```

```python
let_vs_dias.describe()
```

notemos que la desviación estandar (STD), es muy alta en LETALIDAD == 0, y LETALIDAD == 100, esto implica que existen valores extremos que están aumentando la varianza demasiado (outliers...).

Puede implicar posibles registros malos, mal calculo de dias de estancia, u hospitales con muy poca actividad que afectan la muestra.

```python

# 1. Crear bins de LETALIDAD
df_indicadores["letalidad_bin"] = pd.cut(
    df_indicadores["LETALIDAD"],
    bins=[-0.01, 0, 0.01, 0.05, 0.10, 0.20, 1, 100],
    labels=["0%", "0–1%", "1–5%", "5–10%", "10–20%", "20–100%", ">100%"]
)

# 2. Agregar estadísticos
pivot = df_indicadores.groupby("letalidad_bin")["PROMEDIO_DIAS_ESTADA"].agg(["mean", "median", "count"])

# 3. Heatmap (mean)
plt.figure(figsize=(8,4))
sns.heatmap(
    pivot[["mean"]].T,
    annot=True,
    fmt=".2f",
    cmap="YlOrRd"
)

plt.title("PROMEDIO_DIAS_ESTADA por bins de LETALIDAD")
plt.xlabel("LETALIDAD")
plt.show()

# Esto lo hizo la IA porsiacaso...
```

### Establecimientos con más letalidad (proporción de egreso como fallecido del total de egresos)

Por simpleza intentaré otro analisis de letalidad, esta vez agrupando por nombre de los servicios de salud desde la feature "ESTABLECIMIENTO"

```python
df_indicadores.groupby("ESTABLECIMIENTO")["LETALIDAD"].mean()
```

```python
# Saquemos los 10 establecimientos con más letalidad en promedio

top_10_letalidad = df_indicadores.groupby("ESTABLECIMIENTO")["LETALIDAD"].mean().sort_values(ascending=False)
top_10_letalidad.head(10)
```

```python
# ahora usamos ambas agrupoaciones como comparacion en un grafico, top_10_fallecidos y top_10_letal

plt.plot(top_10_letalidad.head(10), label="Letalidad en promedio por hospital")
plt.legend()
plt.xlabel("Establecimiento")
plt.ylabel("Cantidad")
plt.title("Letalidad")

plt.fill_between(top_10_letalidad.head(10).index, top_10_letalidad.head(10).values, alpha=0.3)

# y rotamos los nombres de hospitales porque osino no se ve nada XD
plt.xticks(rotation=90)
plt.show()
```

### Letalidad según año

```python
letalidad_por_ano = df_indicadores.groupby("PERIODO")["LETALIDAD"].mean()
letalidad_por_ano.head(10)
```

```python
plt.plot(letalidad_por_ano)
plt.xlabel("Año")
plt.fill_between(letalidad_por_ano.index, letalidad_por_ano.values, color='skyblue', alpha=0.4)
plt.ylabel("Letalidad promedio")
plt.title("Letalidad ponderada por año")
plt.show()
```

Notamos que la letalidad ponderada por año ha disminuido a niveles inferiores que el 2014... Será esto por falta de datos además? podríamos comparar la gráfica con la cantidad de egresos totales por año...

```python
egresos_totales_year = df_indicadores.groupby("PERIODO")["NUMERO_EGRESOS"].sum()
egresos_totales_year
```

```python
plt.plot(egresos_totales_year)
# el fill lo hizo la IA por cierto... (le dije que queria que quedara bonito abajo de la curva :D)
plt.fill_between(egresos_totales_year.index, egresos_totales_year.values, color='skyblue', alpha=0.4)
plt.xlabel("Año")
plt.ylabel("Cantidad de egresos totales")
plt.title("Egresos Totales por Año")
plt.ylim(bottom=0)
plt.show()
```

Notamos que en el COVID hubieron muchos menos egresados en general de los servicios públicos de salud... Pero no parece haber una correlación muy fuerte en cuanto a la letalidad respectiva.
